Sample Dataset

```python
import pandas as pd

sensus = {
    'tinggi':[158,170,183,191,155,163,180,158,170],
    'jk':['pria','pria','pria','pria','wanita','wanita','wanita','wanita','wanita'],
    'berat':[64,86,84,80,49,59,67,54,67]
}

sensus_df = pd.DataFrame(sensus)
sensus_df
```

Regression dengan KNN

features dan target

```python
import numpy as np

x_train = np.array(sensus_df[['tinggi','jk']]) # features
y_train = np.array(sensus_df['berat']) # target

print("X train : ", x_train)
print("Y train : ", y_train)
```

Preprocess Dataset : Konversi data ke numerik dan biner

```python
x_train_transposed = np.transpose(x_train)

print("X train : ",x_train)
print("X train_transpose : ",x_train_transposed)
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
jk_binarized = lb.fit_transform(x_train_transposed[1])

print("JK", x_train_transposed[1])
print("JK", jk_binarized.flatten())
```

```python
jk_binarized = jk_binarized.flatten()
jk_binarized
```

```python
x_train_transposed[1] = jk_binarized
x_train = x_train_transposed.transpose()

print("X train transpose : ",x_train_transposed[1])
print("X train : ",x_train)
```

Training KNN Regression Model

```python
from sklearn.neighbors import KNeighborsRegressor

k = 3
model = KNeighborsRegressor(n_neighbors=k)
model.fit(x_train, y_train)
```

Prediksi Berat badan

```python
x_new = np.array([[155, 1]])
x_new
```

```python
y_pred = model.predict(x_new)
y_pred
```

Evaluasi KNN Regression Model

```python
x_test = np.array([[168,0],[180, 0], [160, 1],[169,1]])
y_test = np.array([65,96,52,67])

print("X test", x_test)
print("Y test", y_test)
```

```python
y_pred = model.predict(x_test)
y_pred
```

Coefficient of Determination atau r2

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_pred)
print("R2 Score : ",r_squared)
```

Mean Absolute Error | Mean Absolute Deviation

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error Score : ",mean_absolute_error)
```

Mean Squared Error | Mean Squared Deviation

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test, y_pred)
print("Mean Squared Error Score : ",MSE)
```

Standard Score

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
```

```python
x_train = np.array([[1700,0],[1600,1]])
x_train_scaled = ss.fit_transform(x_train)
print("X Train Scaled : \n",x_train_scaled)

x_new = np.array([[1640,0]])
x_new_scaled = ss.transform(x_new)
print("X New Scaled : \n",x_new_scaled)

jarak = [euclidean(x_new_scaled[0],d) for d in x_train_scaled]
print("Jarak", jarak)
```

Menerapkan Features Scaling pada KNN

```python
x_train = np.array([[158, 0],[170, 0],[183, 0],[191, 0],[155, 1],[163, 1],[180, 1],[158, 1],[170, 1]])
y_train = np.array([64, 86 ,84 ,80, 49, 59 ,67 ,54 ,67])

x_test = np.array([[168,0],[180,0],[160,1],[169,1]])
y_test = np.array([65.0, 96.0, 52.0, 67.0])
```

Features Scaling (Standard Scaler)

```python
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)

print("x train scaled", x_train_scaled)
print("x test scaled", x_test_scaled)
```

Training model

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

model.fit(x_train_scaled,y_train)
y_pred = model.predict(x_test_scaled)

MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)

print("MAE", MAE)
print("MSE", MSE)
```
