Load Datasets

```python
import pandas as pd

pizza = {'diameter' : [6,8,10,14, 18],
         'harga' : [7,9,13,17.5,18]}

pizza_df = pd.DataFrame(pizza)
pizza_df
```

Visualisasi Data

```python
import matplotlib.pyplot as plt
pizza_df.plot(kind='scatter',x='diameter',y='harga')
plt.title('Perbandingan harga dan diameter pizza')
plt.xlabel('diameter')
plt.ylabel('harga')
plt.xlim(0,25)
plt.ylim(0,25)
plt.show()
```

Penyesuaian Dataset

```python
import numpy as np
X = np.array(pizza_df['diameter'])
Y = np.array(pizza_df['harga'])

# Reshape Array menjadi 2 dimensi
X = X.reshape(-1,1)
# X
Y
```

Training Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

Visualisasi Simple Linear Regression

```python
X_vis = np.array([0,25]).reshape(-1,1)
Y_vis = model.predict(X_vis)
```

```python
pizza_df.plot(kind='scatter',x='diameter',y='harga')
plt.plot(X_vis,Y_vis,'-r')
plt.title('Perbandingan harga dan diameter pizza')
plt.xlabel('diameter')
plt.ylabel('harga')
plt.xlim(0,25)
plt.ylim(0,25)
plt.show()
```

Mencari nilai slope/koef dan nilai intercept

```python
print('intercept', model.intercept_) # nilai y dimana x=0
print('slope', model.coef_) # Gradient
```

Mencari nilai slope/koef (Manual)

```python
# Ratakan dulu gan
print("X :\n",X)
print("X flatten : ", X.flatten())
print("Y : ", Y)
# Variance
print("\n")
variance_x = np.var(X.flatten(),ddof=1)
print("Variance : ",variance_x)
# Covariance
print("\n")
print("Mencari covariance matrix :\n",np.cov(X.flatten(), Y))
covariance_xy = np.cov(X.flatten(),Y)[0][1]
print("Covariance : ",covariance_xy)
# Mencari nilai slope
print("\n")
slope = covariance_xy / variance_x
print("slope", slope)
```

Mencari nilai intercept (Manual)

```python
intercept = np.mean(Y) - slope * np.mean(X)
print("intercept", intercept)
```

Prediksi Harga Pizza berdasar diameter pizza

```python
diameter_pizza = np.array([12,20,23]).reshape(-1,1)

prediksi_harga = model.predict(diameter_pizza)

for diameter, harga in zip(diameter_pizza, prediksi_harga):
  print("Diameter pizza : ",diameter, " Prediksi Harga : ",harga)
```

Evaluasi Hasil Training Model

```python
# Training & Testing Dataset
X_train = np.array([6,8,10,14,18]).reshape(-1,1)
Y_train = np.array([7,9,13,17.5, 18])

X_test = np.array([8,9,11,16,12]).reshape(-1,1)
Y_test = np.array([11,8.5,15,18, 11])

# Training Simple Linear Regression
model = LinearRegression()
model.fit(X_train, Y_train)

# Evaluasi Linear Regression model dengan Coefficient of Determination atau R(kuadrat)
from sklearn.metrics import r2_score
y_pred = model.predict(X_test)
r_squared = r2_score(Y_test,y_pred)
print("R2 Score :", r_squared)
```

Mencari nilai R(Kuadrat) manual 

```python
# R(Kuadrat) = 1 - SS(residual) / SS(total)
# SS = Sum Squared
# SS(Residual) Selisih antara setiap nilai Y di testing set dengan nilai prediksinya setelah itu nilai selisihnya dipangkatkan 2 dan dijumlahkan secara keseluruhan
ss_res = sum([(y_i - model.predict(x_i.reshape(-1,1))[0])**2 for x_i, y_i in zip(X_test, Y_test)])
print("SS_Res : ", ss_res)
# SS(Total)  Selisinhantara setiap nilai Y di testing set dengan nilai rata ratanya setelah itu nilai selisihnya dipangkatkan 2 dan dijumlahkan secara keseluruhan
mean_y = np.mean(Y_test)
ss_tot = sum([(y_i - mean_y)**2 for y_i in Y_test])
print("SS_Total :", ss_tot)
# Nilai R(Kuadrat)
r_squared = 1 - (ss_res / ss_tot)
print("Nilai R2", r_squared)
```
