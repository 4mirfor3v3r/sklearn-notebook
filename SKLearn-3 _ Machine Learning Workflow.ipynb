Load Datasets

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
Y = iris.target
```

Split Datasets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.4, random_state=1)
```

Training Model

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
```

Evaluasi hasil training

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_test)
acc = accuracy_score(y_pred, y_test)
print("Accuracy :", acc)
```

Pemanfaatan hasil train model

```python
data_baru = [[5,5,3,2],[2,4,3,5]]

preds = model.predict(data_baru)
preds
```

```python
pred_species = [iris.target_names[p] for p in preds]
print("Hasil Prodeiksi", pred_species)
```

Dumping Model machine Learning menjadi file joblib (Export hasil trained model)

```python
import joblib

joblib.dump(model, 'iris_classifier_knn.joblib')
```

Import model hasil train ke production

```python
production_model = joblib.load('iris_classifier_knn.joblib')
```
