Load sample Data

```python
import numpy as np
from sklearn import preprocessing

sample_data = np.array([[2.1, -1.9, 5.5],
                        [-1.5, 2.4, 3.5],
                        [0.5, -7.9, 5.6],
                        [5.9, 2.3, -5.8]
                        ])
sample_data.shape
```

Binarization | Memisah data dalam 2 nilai (diatas dan dibawah threshold)

```python
preprocessor = preprocessing.Binarizer(threshold=0.5)
binarized_data = preprocessor.transform(sample_data)
binarized_data
```

Scaling data | memisah data dalam range tertentu

```python
preprocessor.fit(sample_data)
scaled_data = preprocessor.transform(sample_data)
# combine fit and transform
# scaled_data = preprocessor.fit_transform(sample_data)

scaled_data
```

Normalization L1 Least Absolute Deviation

```python
normalized_l1_data = preprocessing.normalize(sample_data,norm="l1")
normalized_l1_data
```

Normalization L2 Least squares

```python
normalized_l2_data = preprocessing.normalize(sample_data,norm="l2")
normalized_l2_data
```

```python
import matplotlib.pyplot as plt
X = normalized_l1_data

plt.scatter(X[:,0],X[:,1]) # Small
plt.scatter(normalized_l2_data[:,0],normalized_l2_data[:,1]) # middle
plt.scatter(sample_data[:,0],sample_data[:,1]) # large

plt.grid(True)
plt.show
```
